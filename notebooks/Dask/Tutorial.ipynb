```python
from dask import delayed
import numpy as np
from scipy.ndimage import label

    
@delayed
def gen_image():
    return np.random.uniform(-1024, 1024, size = (512, 512, 512))

@delayed
def apply_thresh(in_img, thresh_val):
    return in_img>thresh_val

@delayed
def label_comp(in_thresh):
    return label(in_thresh)[0]

@delayed
def get_max(in_imgs):
    return np.max([c_img.max() for c_img in in_imgs])
```

```python
seed_img = gen_image()
all_mask = [label_comp(apply_thresh(seed_img, c_thresh)) for c_thresh in range(-1000, 1000, 200)]
most_comp = get_max(all_mask)
```

```python
most_comp.visualize()
```
